```python
# text file with car Num as TS XX A-Z XXXX , # of seats , # of bags
import random
import string
file=open("cabdata.txt","w+")
file.write('CarNum,numofseats,numofbags,type,When\n')
typ=['SUV','Hatchback','Sedan']
whn=['Now','Later']
num=['one','two','three','four','five','six','seven','eight']
for i in range(500):
    
    file.write('TS '+str(random.randint(1,100))+" "+ random.choice(string.ascii_uppercase) + random.choice(string.ascii_uppercase) +" "+ str(random.randint(1,10))+ str(random.randint(1,10))+ str(random.randint(1,10))+ str(random.randint(1,10))+","+str(random.randint(2,8))+","+str(random.choice(num))+","+random.choice(typ)+","+random.choice(whn)+'\n')
    
file.close()
```
